- (3 pts) Докажите, что $\| A \|_F \le \sqrt{\mathrm{rank}(A)} \| A \|_2$.

__Доказательство:__

По определению фробениусова норма матрицы: 
$$
  \Vert A \Vert_F \stackrel{\mathrm{def}}{=} \Big(\sum_{i=1}^n \sum_{j=1}^m |a_{ij}|^2\Big)^{1/2} = \sqrt{\mathrm{Tr}(A^{*}A)}
$$

Спектральная норма матрицы:
$$
   \Vert A \Vert_2 = \sigma_1(A) = \sqrt{\lambda_\max(A^*A)}
$$

где $\sigma_1(A)$ – максимальное сингулярное число матрицы $A$. 

Матрица $A^*A$ эрмитова, поэтому диагонализуема в унитарном базисе. $A^*A\succeq 0$ (неотрицательно определена), поэтому собственные значения неотрицательны.
Поэтому существует унитарная матрица $V = [v_1, \dots, v_n]$ такая что

$$
    V^* A^* A V = \text{diag}(\sigma_1^2,\dots, \sigma_n^2), \quad \sigma_1\geq \sigma_2\geq \dots \geq \sigma_n \geq 0.
$$

Более того, если $\text{rank}(A) = r$, тогда $\sigma_{r+1} = \dots = \sigma_n = 0$.

Следовательно, можно записать: 

$\mathrm{Tr}(A^* A) =  \sum \limits_{i=1}^{r} \sigma_{i}^2$

Таким образом, $ ||A||_F = \sqrt{\sum \limits_{i=1}^{r} \sigma_{i}^2} \leq \sqrt{r \cdot \sigma_{1}^2} = \sqrt{\mathrm{rank}(A)} ||A||_2$

__Доказательство:__

То есть, $ \sigma_{1} = \sigma_{2} = ... = \sigma_{k} = \sigma$

Покажем, что такое возможно для матрицы $ A = \sigma U $, где U - унитарная матрица

Таким образом, предположим, что все собственные значения матрицы $A$ равны, тогда все собственные значения эрмитовой матрицы $ A^*A $ также равны и $ A^*A = \sigma^2 I$

То есть, $ \frac{A^*}{\sigma}\frac{A}{\sigma} = I $

$ U = \frac{A}{\sigma} $ - унитарная матрица

Следовательно, $ A = \sigma U $

Существование матрицы для равенства в оценке доказано.

- (5 pts) Для векторов важным дополнительным свойством является количество ненулевых элементов. В задаче решения недоопределённой системы линейных уравнений это свойство является одним из возможных дополнительных условий на решение. Аналогичная задача для матриц – это поиск наилучшего приближения данной матрицы с минимальным рангом. То есть аналогом разреженности для векторов является малоранговость для матриц. Возникает вопрос, какая норма является аналогичной первой векторной нормы? Ответом является так называемая ядерная норма 

$$\|A\|_{nuclear} = \sum_{i=1}^n \sigma_i, $$

где $\sigma_i$ – сингулярные числа матрицы $A$. Покажите, что это выражение действительно является нормой.

__Доказательство:__

- $\Vert \cdot \Vert$ называется матричной нормой, если это векторная норма на векторном пространстве матриц размера $n \times m$:
    1. $\|A\| \geq 0$ и если $\|A\| = 0$ то $A = O$
    3. $\|\alpha A\| = |\alpha| \|A\|$
    4. $\|A+B\| \leq \|A\| + \|B\|$ (неравенство треугольника)
    
Проверим выполнение этих трех условий: 

1. Так как $\sigma_i \geq 0$, следовательно, $\|A\|_{nuclear} = \sum_{i=1}^n \sigma_i \geq 0$, $\|A\|_{nuclear} = 0$, то $\sum_{i=1}^n \sigma_i = 0$, следовательно, $\sigma_1 = \sigma_2 = ... = \sigma_n = 0$, то $\mathrm{rank}(A)=0$, следовательно, $A=0$
2. $A^*A x = \sigma^2 x$, $(\alpha A)^*(\alpha A) x = (\sigma^{'})^2 x$, то есть $(\sigma^{'})^2 = \sigma^2 |\alpha|^2$, $ \sigma^{'} = |\alpha| \sigma$

Таким образом, $|\alpha A\|_{nuclear} = \sum_{i=1}^n \sigma^{'}_i = \sum_{i=1}^n |\alpha| \sigma_i = |\alpha| \sum_{i=1}^n \sigma_i = |\alpha| |A\|_{nuclear}$
3. $\|A + B\|_{nuclear} = \sum_{i=1}^n \sigma_i$

Для начала покажем, что $||A+B||_2 \leq ||A||_2 + ||B||_2$

$||A+B||_2 = \sup \limits_{x \neq 0} \frac{||Ax+Bx||}{||x||} \leq \sup \limits_{x \neq 0} (\frac{||Ax||}{||x||} + \frac{||Bx||}{||x||}) = ||A||_2 + ||B||_2$

Так как $||A||_2 = \sigma_{1}(A)$, $||B||_2 = \sigma_{1}(B)$, то $\sigma_{1}(A+B) \leq \sigma_{1}(A) + \sigma_{1}(B)$ 

Следовательно, $\|A+B\|_{nuclear}  = \sum_{i=1}^n \sigma_i(A+B) \leq \sum_{i=1}^n \sigma_i(A) + \sum_{i=1}^n \sigma_i(B) = \|A\|_{nuclear} + \|B\|_{nuclear}$

Все условия на норму выполнены.

- (1 pts) Покажите, что если матрица $A$ обратима, то $\|A^{-1}\|_F \geq \frac{\sqrt{n}}{\|A\|_F}$

__Доказательство:__

- По определению обратной матрицы: $A^{-1} A =I$

$|| A^{-1} A||_F = ||I||_F = \sqrt{n}$

Докажем субмультипликативность фробениусовой нормы: $||AB||_F \leq ||A||_F ||B||_F$ 

__Утверждение 1.__ Если $\alpha_1 \geq \alpha_2 \geq ... \geq \alpha_n $ - сингулярные числа матрицы $A$, $\beta_1 \geq \beta_2 \geq ... \geq \beta_n $ - сингулярные числа матрицы $B$, тогда $|\mathrm{Tr}(AB)| \leq \sum \limits_{i=1}^{n} \alpha_i\beta_i$.

__Доказательство:__

Действительно, пусть $A=U_1\Sigma_1V_1^*, B=U_2\Sigma_2V_2^*$ - сингулярные разложения матриц. Тогда $\mathrm{Tr}(AB) = \mathrm{Tr}(U_1\Sigma_1V_1^*U_2\Sigma_2V_2^*) = $(круговое перестановочное свойство следа матрицы) $= \mathrm{Tr}(V_2^*U_1\Sigma_1V_1^*U_2\Sigma_2)$ = \{ $U = U_1^*V_2, V=V_1^*U_2$\} $=\mathrm{Tr}(U^*\Sigma_1V\Sigma_2) = |\sum u_{ij} v_{ij}\alpha_i\beta_j| \leq \frac{|u_{ij}|^2\alpha_i\beta_j + |v_{ij}|^2\alpha_i\beta_j}{2} = $ \{ $ \sum u_{ij} = 1, \sum v_{ij} = 1$\} $\leq \frac{\alpha_i\beta_j + \alpha_i\beta_j}{2} = \sum \alpha_i\beta_j$

__Утверждение 2.__

Если A и B - эрмитовы матрицы, то $|\mathrm{Tr}(AB)| \leq \mathrm{Tr}(A) \mathrm{Tr}(B)$

__Доказательство:__

У эрмитовых матриц совпадают собственные значения с сингулярными числами, как было показано выше, поэтому, учитывая утверждение 1, получим, что:

$|\mathrm{Tr}(AB)| \leq \sum \alpha_i\beta_i \leq \sum \alpha_i \sum \beta_i = \mathrm{Tr}(A) \mathrm{Tr}(B)$, где $\alpha_i, \beta_i$ - собственные значения матриц A и B

Тогда, учитывая 2 верхних утверждения,

$||AB||_F = \mathrm{Tr}(B^*A^*AB) = \mathrm{Tr}(BB^*A^*A) \leq \mathrm{Tr}(BB^*)\mathrm{Tr}(A^*A) = \mathrm{Tr}(B^*B)\mathrm{Tr}(A^*A)= ||A||_F ||B||_F$ (воспользовались круговым перестановочным свойством следа матрицы)

То есть, $|| A^{-1} A||_F \leq ||A^{-1}||_F ||A||_F$

Следовательно, $||A^{-1}||_F \geq \frac{\sqrt{n}}{||A||_F}$

__Решение__

Матрица обратима тогда и только тогда, когда она невырожденная.

Таким образом, найдем определитель матрицы. 

Проделаем над матрицей серию ЭПС: а именно, будем поочередно вычитать из последней строки первую, помноженную на $a_1$, затем вторую, помноженную на $a_2$ и т.д. Таким образом, получим: 

$A = \begin{bmatrix} I & a \\ a^{\top} & 0 \end{bmatrix} \sim \begin{bmatrix} I & a \\ 0...0 & -a_1^2 -a_2^2-...-a_n^2 \end{bmatrix}$

где $a_1, a_2, ..., a_n$ -- координаты вектора $a$

Определитель верхнетреугольной матрицы: $\det(A) = -a_1^2 -a_2^2-...-a_n^2 = -a^T \cdot a \neq 0$ 

Откуда следует, что $a \neq 0$, что является условием обратимости матрицы.

Найдем обратную матрицу: 

Для этого сначала диагонализуем матрицу, ведь для диагональной матрицы обратной является диагональная матрица с обратными элементами на диагонали. 

Симметричная матрица всегда диагонализуема, причем на диагонали стоят собственные значения с учетом их кратности.

````python
# INPUT : diag_broadcast - list of diagonals value to broadcast, length equal to 3 or 5; n - integer, band matrix shape.
# OUTPUT : L - 2D np.ndarray, L.shape[0] depends on bandwidth, L.shape[1] = n-1, do not store main diagonal, where all ones;                  add zeros to the right side of rows to handle with changing length of diagonals.
#          U - 2D np.ndarray, U.shape[0] = n, U.shape[1] depends on bandwidth;
#              add zeros to the bottom of columns to handle with changing length of diagonals.
def band_lu(diag_broadcast, n):
    # enter your code
    x = int(len(diag_broadcast)/2)
    if len(diag_broadcast) == 3:
        p = 1 # полуширина матрицы
    else:
        p = 2
    l = np.eye(n)
    u = np.zeros((n, n))
    
    for i in range(n):
        # так матрица трех/пяти-диагональная, то можно проходить не все элементы, а только заведомо ненулевые
        bot_border = np.maximum(0, i - p)
        top_border = np.minimum(n - 1, p + i) + 1

        for j in range(i, top_border):
            u[i, j] = diag_broadcast[j - i + x]  - np.matmul(l[i, bot_border:i],(u[bot_border:i, j]))
    
        for k in range(i+1, top_border): 
            l[k, i] = (diag_broadcast[i - k + x] - np.matmul(l[k, bot_border:i],(u[bot_border:i, i]))) / u[i, i]
            
    # записываем диагонали
    res_l = np.zeros((p, n-1)) 
    res_u = np.zeros((n, p+1))
    for i in range(-1, -1-p, -1):
        res_l[1+i, :n+i] = np.diag(l, k=i)
    for i in range(0, p+1):
        res_u [:n-i, i] = np.diag(u, k=i)  
    return res_l, res_u
````

````python
# замеряем время работы band_lu на матрице размера от n*n, n от 4 до 1000
time_band_lu = []
for i in range(4, 1000):
    arr = sp.diags([3, -5, 4], [-1, 0, 1], shape=(i, i)).toarray() 
    start_time = time.time()
    band_lu([3, -5, 4], i)
    time_band_lu.append(time.time() - start_time)

# замеряем время работы встроенной функции scipy lu на матрице размера от n*n, n от 4 до 1000
time_lu = []
for i in range(4, 1000):
    arr = sp.diags([3, -5, 4], [-1, 0, 1], shape=(i, i)).toarray()    
    start_time = time.time()
    la.lu(arr, permute_l=True)
    time_lu.append(time.time() - start_time)
````

````python
# строим зависимость времени работы алгоритмов от размеров матриц
x = range(4, 1000)

plt.plot(x, time_band_lu, 'r', label = 'band_lu')
plt.plot(x, time_lu, 'b', label = 'LU from scipy')
plt.title('Зависимость времени работы алгоритма от размеров матрицы')
plt.xlabel('Размер матрицы n')
plt.ylabel('Время работы алгоритма')
plt.legend(shadow = False,fontsize = 10)
plt.show()
````

При учете ленточной структуры мы отбрасываем очень много вычислительных операций, ведь в такой матрице много нулей. Чем больше размер матрицы, тем меньше мы проделываем арифметических операций по отношению к алгоритму, не учитывающему структуру матрицы. А именно, каждый шаг исключения занимает порядка $p^2$ операций, так что общая сложность будет $O(p^2n)$, при константном $p$ это обеспечит линейную сложность в отличии от $O(n^3)$ кубической сложности в общем случае.

## Задача 3:  QR разложение (27 pts)

- (5 pts) Реализуйте метод Грама-Шмидта для вычисления QR разложения. На вход подаётся прямоугольная матрица, функция возвращает матрицы ```Q``` и ```R``` из QR разложения. Прототип приведён ниже и не должен изменяться 

````python
# INPUT : rectangular matrix A
# OUTPUT: matrices Q - orthogonal and R - upper triangular such that A = QR
def gram_schmidt_qr(A):
    # your code is here
    n = A.shape[0]
    m = A.shape[1]
    Q = np.zeros((n,m))
    R = np.zeros((n,m))
    u = A[:,0]
    Q[:,0] = u/np.linalg.norm(u)
    for k in range(1, m):
        sum_ = 0
        for i in range(0, k):
            sum_ += np.dot(A[:,k], Q[:,i])*Q[:,i]
        u = A[:, k] - sum_ 
        Q[:, k] = u/np.linalg.norm(u)
    for i in range(n):
        for j in range(i, m):
            R[i, j] = np.dot(A[:, j], Q[:, i])
    return Q, R
````

- (5 pts) Реализуйте модифицированный метод Грама-Шмидта для вычисления QR разложения. На вход подаётся прямоугольная матрица, функция возвращает матрицы Q и R из QR разложения. Прототип приведён ниже и не должен изменяться

````python
# INPUT : rectangular matrix A
# OUTPUT: matrices Q - orthogonal and R - upper triangular such that A = QR
def modified_gram_schmidt_qr(A): # 5 pts
    # your code is here
    n = A.shape[0]
    m = A.shape[1]
    Q = np.zeros((n,m))
    R = np.zeros((n,m))
    u = A[:,0]
    Q[:,0] = u/np.linalg.norm(u)
    for j in range(1, m):
        u = A[:, j]
        for i in range(0, j):
            R[i,j] =  np.dot(A[:,j], Q[:,i])
            u -= R[i,j]*Q[:,i]
        Q[:, j] = u/np.linalg.norm(u)
        R[j,j] =  np.linalg.norm(u)    
    return Q, R
````

- (7 pts) Реализуйте алгоритм вычисления QR разложения, основанный на преобразованиях Хаусхолдера. На вход подаётся прямоугольная матрица, функция возвращает матрицы Q и R из QR разложения. Прототип приведён ниже и не должен изменяться. Обратите внимание на эффективность реализации операции умножения матриц Хаусхолдера для получения матрицы ```Q```.

````python
# INPUT : rectangular matrix A
# OUTPUT: matrices Q - orthogonal and R - upper triangular such that A=QR
def householder_qr(A): # 7 pts
    # your code is here
    n = A.shape[0]
    m = A.shape[1]
    R = A.copy()
    H_res = np.eye(n)

    for j in range(0, m):
        x = R[j:, j].reshape(-1, 1)

        if len(x) != 0:
            v = x + np.copysign(np.linalg.norm(x), x[0]) 

        H = np.eye(n)
        H[j:, j:] = H[j:, j:] - 2/np.linalg.norm(v)**2*np.matmul(v, v.T)
        R = np.matmul(H, R)
        H_res = np.matmul(H, H_res)
    Q = H_res.T
    R = np.triu(R)    
    return Q, R
````

Проблема могла возникнуть, если x был бы почти коллинеарен $e_1$, то вектор отражения $v = x − ||x||_2e_1$ имеет бы малую норму, поэтому $v$ выбирается следующим образом: $v = x + sign(x_1)||x||^2e_1$, где $e_1$ - единичный вектор. Таким образом, $||v||^2 ≥ ||x||^2$, что позволяет получить практически идеальную ортогональность.

- (5 pts) Пусть дано QR разложение матрицы $A$ размера $m \times n, \; m > n$. К матрице $A$ добавили ещё одну строку после последней, то есть теперь есть матрица $\hat{A} = \begin{bmatrix} A \\ a \end{bmatrix}$ размера $(m+1) \times n$. Покажите, как найти QR разложение матрицы $\hat{A}$, используя QR разложения матрицы $A$. Сравните асимптотические сложности предложенного вами способа и метода, который не использует QR разложение матрицы $A$. Также экспериментально покажите, как время работы этих двух методов зависит от размерности по числу строк и столбцов матрицы.

__Решение:__

Для того чтобы построить, QR-разложение этой матрицы, используем ее блочную структуру. 
Строка a имеет вырожденное разложение: $q=1, R=a$.
Тогда построим следующее блочное равенство: $\begin{bmatrix} A \\ a \end{bmatrix} = \begin{bmatrix} QR \\ 1 \cdot a \end{bmatrix} = \begin{bmatrix} Q & 0\\ 0 & 1 \end{bmatrix} \begin{bmatrix} Q^{'} \end{bmatrix} \begin{bmatrix} R^{'} \\ 0 \end{bmatrix}$, где $Q^{'}R^{'}$ - разложение матрицы $\begin{bmatrix} R \\ a \end{bmatrix}$

Таким образом, итоговое разложение: $\hat{A} = \hat{Q}\hat{R}$, где $\hat{Q} = Q^{''}Q^{'}, R=R^{'}$, $Q^{''} = \begin{bmatrix} Q & 0\\ 0 & 1 \end{bmatrix}$

Данный подход позволяет существенно уменьшить сложность алгоритма, ведь матрица $\begin{bmatrix} R \\ a \end{bmatrix}$, которую мы будем раскладывать практически верхнетреугольная (внизу еще есть новая строка), таким образом, если в методе QR-разложения использовать информацию о большом количестве нулевых элементов, можно достичь хорошего времени работы. 
Для этого построено QR-разложение с ортогонализацией Грама-Шмидта, которое при ортогонализации столбцов перемножает только ненулевые элементы.
Асимптотическая сложность QR-разложения в общем случае: $O(n^2m)$. В случае разбиения на блоки и учитывания нулевых элементов: число арифметических операций примерно равно: $\frac{n^2+n}{2}m - \frac{m^2+m}{2}$, что при больших значениях $m$, позволяет выйти практически на $O(nm)$.

````python
# алгоритм QR-разложения с ортогонализацией Грама-Шмидта, учитывающий наличие нулевых элементов
def gram_schmidt_qr_new(A):
    n = A.shape[0]
    m = A.shape[1]
    Q = np.zeros((n,m))
    R = np.zeros((n,m))
    u = A[:,0]
    Q[:,0] = u/np.linalg.norm(u)
    for k in range(1, m):
        sum_ = 0
        for i in range(0, k):
            temp_a = A[:,k]
            temp_q = Q[:,i]
            non_z = np.nonzero(temp_a)
            a_1 = temp_a[non_z]
            a_2 = temp_q[non_z]
            sum_ += np.dot(a_1, a_2)*Q[:,i]
        u = A[:, k] - sum_ 
        Q[:, k] = u/np.linalg.norm(u)
    for i in range(n):
        for j in range(i, m):
            R[i, j] = np.dot(A[:, j], Q[:, i])
    return Q, R
````

````python
def qr_with_new_row(Q, R, a):
    C = np.vstack((R, a))
    q_c, r_c = gram_schmidt_qr_new(C)
    q = np.vstack((Q, np.zeros(Q.shape[1])))
    q = np.hstack((q, np.zeros((q.shape[0],1))))
    q[-1, -1] = 1
    q_res = np.matmul(q, q_c)
    return q_res, r_c
````

## Задача 4. Сжатие полносвязных слоёв в нейросети (30 pts)

В этой задаче вам предлагается исследовать, как изменится точность предсказания нейросети, которая классифицирует изображения, если вместо полносвязных слоёв использовать их сжатое представление, полученное с помощью SVD. Результатом решения этой задачи должен быть график зависимости точности классификации на тестовой выборке от количества параметров нейронной сети.

#### Сначала скачаем данные CIFAR10

````python
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision import datasets, transforms
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

batch_size = 100

transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

train_loader = torch.utils.data.DataLoader(datasets.CIFAR10('./', train=True, download=True, transform=transform), 
                                        batch_size=batch_size, shuffle=True)

test_loader = torch.utils.data.DataLoader(datasets.CIFAR10('./', train=False, transform=transform), 
                                          batch_size=batch_size, shuffle=True)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
````

````python
def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.figure(figsize=(20, 10))
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images
dataiter = iter(train_loader)
images, labels = dataiter.next()

# show images
imshow(torchvision.utils.make_grid(images))
# print labels
print(' '.join('%5s' % classes[labels[j]] for j in range(8)))
````

### Архитектура нейронной сети

Для простоты и демонстрации принципа сжатия слоёв нейронных сетей рассмотрим архитектуру, состоящую только из полносвязных слоёв (то есть линейных преобразований) и стандартных ReLU-нелинейностей. Опять же для наглядности выбрали размерность 1000 для полносвязных слоёв. 

````python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.fc2 = nn.Linear(1000, 1000)
        self.fc3 = nn.Linear(1000, 1000)
        self.fc4 = nn.Linear(1000, 1000)
        self.fc5 = nn.Linear(1000, 1000)
        self.fc6 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.fc2(x)
        x = self.ReLU(x)
        x = self.fc3(x)
        x = self.ReLU(x)
        x = self.fc4(x)
        x = self.ReLU(x)
        x = self.fc5(x)
        x = self.ReLU(x)
        x = self.fc6(x)
        return F.log_softmax(x, dim=1)
````

#### Функции для обучения и тестирования результата после каждого прохождения всех данных 

````python
def train(model, train_loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
#             data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
    
    return 100. * correct / len(test_loader.dataset)
````

### Зададим параметры обучения и вывода промежуточных результатов

````python
log_interval = 50
epochs = 7
````

````python
model = Net()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(1, epochs + 1):
    train(model,  train_loader, optimizer, epoch)
    test(model, test_loader)
````

Сингулярные числа убывают быстро (до примерно 800-го экспоненциально), так что, взяв, в усеченном SVD-разложении даже небольшое r, не потеряется большее количество информации.

````python
def svd(x, r):
    u, s, v = torch.svd(x.data[:])
    u1 = u[:, :r]
    s1 = s[:r]
    v1 = v[:, :r]
    return u1, s1, v1
````

````python
# модель для сжатия всех полносвязных слоев с помощью SVD разложения
class Net_new(Net):
    def __init__(self, mod, r):
        super(Net_new, self).__init__()
        self.mod = mod
        self.r = r 
        
        # производится сжатие полносвязных слоев
        self.fc2_u, self.fc2_s, self.fc2_v = svd(self.mod.fc2.weight, self.r)
        self.fc3_u, self.fc3_s, self.fc3_v = svd(self.mod.fc3.weight, self.r)
        self.fc4_u, self.fc4_s, self.fc4_v = svd(self.mod.fc4.weight, self.r)
        self.fc5_u, self.fc5_s, self.fc5_v = svd(self.mod.fc5.weight, self.r)
   
    # эффективная реализация с учетом truncated-svd
    def forward(self, x):
        x = self.mod.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        
        w_new = torch.matmul(torch.matmul(self.fc2_u, torch.diag_embed(self.fc2_s)), self.fc2_v.transpose(-2, -1))
        x = torch.matmul(x, w_new.t()) + self.mod.fc2.bias
        x = self.ReLU(x)
        
        w_new = torch.matmul(torch.matmul(self.fc3_u, torch.diag_embed(self.fc3_s)), self.fc3_v.transpose(-2, -1))
        x = torch.matmul(x, w_new.t()) + self.mod.fc3.bias
        x = self.ReLU(x)
        
        w_new = torch.matmul(torch.matmul(self.fc4_u, torch.diag_embed(self.fc4_s)), self.fc4_v.transpose(-2, -1))
        x = torch.matmul(x, w_new.t()) + self.mod.fc4.bias
        x = self.ReLU(x)
        
        w_new = torch.matmul(torch.matmul(self.fc5_u, torch.diag_embed(self.fc5_s)), self.fc5_v.transpose(-2, -1))
        x = torch.matmul(x, w_new.t()) + self.mod.fc5.bias
        x = self.ReLU(x)
        
        x = self.mod.fc6(x)
        return F.log_softmax(x, dim=1)
````

````python
# замер времени вычисления от ранга сжатия
import timeit
ac = []
time_ = []
for r in range(1, 101, 5):
    mod = Net_new(model, r)
    t = %timeit -o -q -n 3 -r 1 test(mod, test_loader)#из-за не очень мощных вычислительных ресурсов взяты небольшие параметры r и n
    ac.append(test(mod, test_loader))
    time_.append(t.average)
````

Чем больше ранг сжатия, тем больше информации о весах мы сохраняем. Однако сингулярные числа убывают очень быстро экспоненциально и достаточно небольшого их количества, чтобы выйти на константную точность. На данном графике, что при числе параметров $\approx 3300000-3400000$ точность выходит на плато, что соответсвует $r \approx 30-40$.

График зависимости времени вычисления от ранга сжатия получился не очень качественный ввиду малых значений параметров timeit, но закономерность проследить можно: время вычисления увеличивается с увеличением ранга сжатия, ведь чем больше ранг, тем большие матрицы необходимо перемножать в функции forward.

### Fine-tuning

- (10 pts) Поскольку в вашей модели, которая использует сжатое представление линейных слоёв, параметрами являются матрицы, полученные из SVD соответствующих линейных слоёв, то их также можно оптимизировать. Эта процедура называется fine-tuning. Предлагается провести обучение этих матриц в течение нескольких эпох (от 1 до 5 будет достаточно) и сравнить насколько выросло качество по сравнению с первоначальными значениями. Покажите значения точности на тестовой выборки и объясните полученные результаты.
